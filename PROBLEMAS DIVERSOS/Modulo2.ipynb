3. A. Escribe un programa que identifique si un número es primo o no lo es:

```python
from math import sqrt
Num=int(input("ingrese un numero: "))
print(Num,":")
def divisores(Num):
    if Num < 2: return "no es primo"
    raiz=int(sqrt(Num))
    cont=0
    for i in range(1,raiz+1):
        if Num%i==0: cont+=1
    return "es primo" if cont==1 else "no es primo"
divisores(Num)
```

3. B. Escribe un programa que encuentre la suma de todos los números primos menores que 100.

```python
from math import sqrt

primos=list()
def divisores(Num):
    if Num < 2: return "no es primo"
    raiz=int(sqrt(Num))
    cont=0
    for i in range(1,raiz+1):
        if Num%i==0: cont+=1
    if cont==1: primos.append(Num)

def mostrarPrimos():
    cont1=0
    for i in primos:
        cont1+=i
    print(primos)
    print("suma de primos: ",cont1)

for i in range(1,100): divisores(i)
mostrarPrimos()
```

8. Escribe un programa que encuentre todos los números perfectos menores que 1000. Un número perfecto es un número entero positivo que es igual a la suma de sus divisores propios positivos (excluyendo el propio número).

```python
from math import sqrt
    
def divisores(Num):
    raiz=int(sqrt(Num))
    cont=1
    for i in range(2,raiz+1):
        if Num%i==0: 
            cont+=i
            if i != Num //i: cont+=Num//i
    return cont

def NumPerfectos(lim):
    perfectos = []
    for e in range(2, lim):
        if divisores(e) == e:
            perfectos.append(e)
    return perfectos

Num=int(input("ingrese un numero: "))
NumPerfectos(Num)
```

```python
registro=[]

for e in range(3):
    alumno=input(f"Ingrese el nombre del alumno {e+1}: ")
    notas=[]
    for i in range(3):
        nota=input(f"Ingrese nota {i+1}: ")
        notas.append(nota)
    registro.append([alumno,notas])

for alumno, notas in registro:
    print(f"Nombre: {alumno}")
    for n in notas:
        print(f"Nota: {n}")
```

2. Definir una función que dado un listado de alumnos evalúe cuántos aprobaron y cuántos desaprobaron, teniendo en cuenta que se aprueba con 4. La nota será el promedio de las 3 notas para cada alumno.

```python
registro=[]
alumnos_f=int(input("Ingrese el numero de alumnos"))
notas_c=int(input("Ingrese el numero de notas"))
for e in range(alumnos_f):
    alumno=input(f"Ingrese el nombre del alumno {e+1}: ")
    notas=[]
    for i in range(notas_c):
        nota=float(input(f"Ingrese nota {i+1}: "))
        notas.append(nota)
    registro.append([alumno,notas])

for alumno, notas in registro:
    print(f"Nombre: {alumno},notas: {notas} -> ", "APROBADO" if sum(notas)/notas_c >=4 else "DESAPROBADO")
```

5. Realizar una función que permita buscar un alumno por nombre, siendo el nombre completo o parcial, y devuelva una lista con los n alumnos que concuerden con ese nombre junto con todos sus datos, incluido el promedio de sus notas.

```python
registro = [
    ["Carlos Pérez", [15, 18, 17]],
    ["Carla Gómez", [12, 14, 16]],
    ["Pedro Ruiz", [19, 20, 18]]
]
nAlumnos=3

def buscarAlumno(registro,nombreBuscar,nAlumnos):
    resultados = []
    for alumno, notas in registro:
        if nombreBuscar.lower() in alumno.lower():
            prom = round(sum(notas)/nAlumnos, 2)
            resultados.append({"nombre": alumno,"notas": notas,"promedio": prom})
    return resultados

nombre = input("Ingrese el nombre o parte del nombre a buscar: ")
coincidencias = buscarAlumno(registro, nombre, nAlumnos)

if coincidencias:
    print("Resultados encontrados:")
    for alumno in coincidencias:
        print(f"Nombre: {alumno['nombre']}, Notas: {alumno['notas']} -> Promedio: {alumno['promedio']}")
else:
    print("No se encontraron coincidencias.")
```
